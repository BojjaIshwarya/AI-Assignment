```python
class State:
    def __init__(self, Rabbit_Leap):
        self.Rabbit_Leap = Rabbit_Leap
    def goalTest(self):
        if self.Rabbit_Leap == ['R', 'R', 'R', '_', 'L', 'L', 'L']:
            return True
        else:
            return False
    def moveGen(self):
        children = []
        i = self.Rabbit_Leap.index('_') 
        if i - 1 >= 0 and self.Rabbit_Leap[i - 1] == 'L':
            new_Rabbit_Leap = self.Rabbit_Leap[:]
            new_Rabbit_Leap[i], new_Rabbit_Leap[i - 1] = new_Rabbit_Leap[i - 1], new_Rabbit_Leap[i]
            children.append(State(new_Rabbit_Leap))
        if i - 2 >= 0 and self.Rabbit_Leap[i - 2] == 'L' and self.Rabbit_Leap[i - 1] == 'R':
            new_Rabbit_Leap = self.Rabbit_Leap[:]
            new_Rabbit_Leap[i], new_Rabbit_Leap[i - 2] = new_Rabbit_Leap[i - 2], new_Rabbit_Leap[i]
            children.append(State(new_Rabbit_Leap))
        if i + 1 < 7 and self.Rabbit_Leap[i + 1] == 'R':
            new_Rabbit_Leap = self.Rabbit_Leap[:]
            new_Rabbit_Leap[i], new_Rabbit_Leap[i + 1] = new_Rabbit_Leap[i + 1], new_Rabbit_Leap[i]
            children.append(State(new_Rabbit_Leap))
        if i + 2 < 7 and self.Rabbit_Leap[i + 2] == 'R' and self.Rabbit_Leap[i + 1] == 'L':
            new_Rabbit_Leap = self.Rabbit_Leap[:]
            new_Rabbit_Leap[i], new_Rabbit_Leap[i + 2] = new_Rabbit_Leap[i + 2], new_Rabbit_Leap[i]
            children.append(State(new_Rabbit_Leap))
        return children
    def __str__(self):
        return self.Rabbit_Leap
    def __eq__(self, other):
        return self.Rabbit_Leap == other.Rabbit_Leap
    def __hash__(self):
        return hash(tuple(self.Rabbit_Leap))
def reconstructPath(goal_node_pair, CLOSED):
    parent_map = {}
    for node, parent in CLOSED:
        parent_map[node] = parent
    path = []
    goal_node, parent = goal_node_pair
    path.append(goal_node)
    while parent:
        path.append(parent)
        parent = parent_map[parent]
    return path
def removeSeen(children, OPEN, CLOSED):
    open_nodes = [node for node, parent in OPEN]
    closed_nodes = [node for node, parent in CLOSED]
    new_nodes = [c for c in children if c not in open_nodes and c not in closed_nodes]
    return new_nodes
def bfs(start):
    OPEN = [(start, None)]
    CLOSED = []
    while OPEN:
        node_pair = OPEN.pop(0)
        N, parent = node_pair
        if N.goalTest():
            print("Goal is found (BFS)")
            path = reconstructPath(node_pair, CLOSED)
            path.reverse()
            for node in path:
                print(node.Rabbit_Leap)
            return
        else:
            CLOSED.append(node_pair)
            children = N.moveGen()
            new_nodes = removeSeen(children, OPEN, CLOSED)
            new_pairs = [(node, N) for node in new_nodes]
            OPEN += new_pairs
    return []
def dfs(start):
    OPEN = [(start, None)]
    CLOSED = []
    while OPEN:
        node_pair = OPEN.pop(0)
        N, parent = node_pair
        if N.goalTest():
            print("Goal is found (DFS)")
            path = reconstructPath(node_pair, CLOSED)
            path.reverse()
            for step in path:
                print(step.Rabbit_Leap)
            return
        else:
            CLOSED.append(node_pair)
            children = N.moveGen()
            new_nodes = removeSeen(children, OPEN, CLOSED)
            new_pairs = [(node, N) for node in new_nodes]
            OPEN = new_pairs + OPEN
    return []
Rabbit_Leap = ['L', 'L', 'L', '_', 'R', 'R', 'R']
start_state = State(Rabbit_Leap)
bfs(start_state)
dfs(start_state)
```
